```
#Mision 1
zones =(gpd.read_file('zonas.shp', crs={'init':'epsg:4326'}))
zones.head()
```

```
# Continuación mision 1
zones.plot()
```

```
#Mision 2
stops =(gpd.read_file('paradas.shp', crs={'init':'epsg:4326'}))
stops.head()
```

```
plt.figure(figsize=(10,6))
ax = plt.gca()
zones.plot(ax = ax, color = 'lightblue',edgecolor='black')
stops.plot(ax = ax, color = 'black', markersize = 10)
```

```
#Mision 3
stopsINzones = gpd.sjoin(stops, zones, op='intersects')
plt.figure(figsize=(10,6))
ax = plt.gca()
zones.plot(ax = ax, color = 'lightblue', edgecolor='black')
stopsINzones.plot(ax=ax, color = 'black', markersize = 10)
```
